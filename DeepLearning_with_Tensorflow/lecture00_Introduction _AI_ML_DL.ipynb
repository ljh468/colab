```python
# 손실함수(MSE)의 변화를 시각화 하기 위한 Utility function
import matplotlib.pyplot as plt
def plot1(data, label, title, xlabel, ylabel):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(data, label=label)
    plt.legend(loc='best')
    plt.grid()
    plt.show()

def plot2(data1, label1, data2, label2, title, xlabel, ylabel):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(data1, label=label1)
    plt.plot(data2, label=label2)
    plt.legend(loc='best')
    plt.grid()
    plt.show()
```

# **예제 1 (Simple Regession)**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
import numpy as np

# Step 1 : 데이터 지정
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([3, 4, 5, 6, 7, 8])

# Step 2 : 모델 구축
model = Sequential()
model.add(Dense(1,input_shape=(1,)))

# Step 3 : 모델 컴파일
model.compile(optimizer=SGD(learning_rate=0.01), loss='mse')
model.summary() # 모델 구조 요약

# Step 4 : 모델 학습 (1000번)
hist = model.fit(x, y, epochs=1000)
# weight와 bias 확인
# weights = model.layers[0].get_weights()[0]
# biases = model.layers[0].get_weights()[1]
# print('weights = {}\tBiases={}'.format(weights, biases))

# Step 5 : 모델 평가 및 예측
input_x = np.array([-3.1, 3.0, 3.5, 15, 20.1, 10000])
predicted = model.predict(input_x)
# print(result)
print('Predicted\tInput')

for i in range(len(predicted)):
    print(predicted[i], '\t', input_x[i]+2)
```

```python
import numpy as np
import pandas as pd
# 손실함수 시각화
plot1(hist.history['loss'], 'train loss', 'Loss','epochs','Loss')
```

# **예제 2 (Multiple regression)**

```python
from tensorflow.keras.optimizers import SGD, Adam

# Step 1 : 데이터 생성
x = np.array([ [1,2,0],[5,4,3],[1,2,-1], [3,1,0],[2,4,2],
               [4,1,2],[-1,3,2],[4,3,3],[0,2,6],[2,2,1],
               [1,-2,-2],[0,1,3],[1,1,3],[0,1,4],[2,3,3] ])
y = np.array( [-4,4,-6,3,-4,
                9,-7,5,6,0,
                4,3,5,5,1])

# Notice y = 2 * 1 - 3 * 2 + 2 * 3
# from_y = [2*data[0] - 3*data[1] + 2*data[2] for data in x]
# print('from_y : ', from_y)

print('x shape : ', x.shape, 'y shape : ', y.shape)

# Step 2 : 모델구축
hidden_units = 10
input_size = 3

model = Sequential()
Dense(hidden_units, activation = 'relu', input_dim = input_size),

model.add(Dense(1, input_shape=(3, ), activation = 'linear'))

# Step 3 : 모델 컴파일
model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')
model.summary()

# Step 4 : 모델 학습
hist = model.fit(x, y, epochs = 1000)

# weight 와 blias 확인
weights = model.layers[0].get_weights()[0]
biases = model.layers[0].get_weights()[1]
print('weights = {}#tbiases={}'.format(weights, biases))

# Step 5 : 모델 평가 및 예측

test_data=[ [5,5,0],[2,3,1],[-1,0,-1],[10,5,2],[4,-1,-2] ]      
true_y = [2*data[0]-3*data[1]+2*data[2] for data in test_data]
predicted_y = model.predict(np.array(test_data))

print('Predicted y\tTrue y')

for i in range(len(predicted_y)):
    print(predicted_y[i],'\t',true_y[i])
```

# **MNIST DataSet**

MNIST는 간단한 컴퓨터 비전 데이터 집합입니다. 다음과 같이 손으로 쓴 자릿수의 이미지로 구성됩니다.

image.png

= 교육 데이터를 위해 손으로 쓴 60000개의 이미지, 테스트 데이터에 따라 다름

교육 데이터(mnist.train)를 위해 손으로 쓴 60,000개의 이미지,<br>
테스트 데이터에 대해 손으로 쓴 자릿수 10,000개의 이미지(mnist.test)<br>
각 이미지는 가로 28픽셀, 세로 28픽셀<br>
우리는 이것을 큰 숫자의 배열로 해석함

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# mnist data 로딩
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Original 데이터 확인
print('x_train shape : {}\ty_train shape : {}\n'.format(x_train.shape, y_train.shape))
print('x_test shape : {}\ty_train shape : {}\n'.format(x_test.shape, y_test.shape))

np.set_printoptions(linewidth=600, precision=2) # linewidth = 라인 최대 글자수, precision = 소수점이하 자리수

# Browsing Original Dataset
print('\n', y_train[0:9])
print('\n', x_train[0])
print('\n', x_train[5])
```

```python
fig = plt.figure(figsize=(6, 6)) # width, height in Inches.
fig.subplots_adjust(hspace=0.7)

for index in range(25):
    plt.subplot(5, 5, index + 1) # 5행 5열
    plt.imshow(x_train[index], cmap='gray')
    plt.axis('off')
    plt.title(str(y_train[index]))

plt.show()

plt.imshow(x_train[9], cmap='gray')
plt.colorbar()
plt.show()
```

```python
import time
start = time.time() # 실행시간 측정을 위해 시작 시간 저장

# Step 1 : 데이터준비, 전처리 가공, 하이퍼파라미터 지정
(x_train, y_train), (x_test, y_test) = mnist.load_data()
image_size = x_train.shape[1]
num_classes = 10
input_size= image_size*image_size
# 데이터 전처리 :

# 1) Dense Layers의 input은 1차원 텐서만 가능
x_train = np.reshape(x_train, [-1, input_size])
x_test = np.reshape(x_test, [-1, input_size])
print('x_train shape : ', x_train.shape)

# 2) Data normalization : # 학습을 더 빨리하고 Local optimum에 빠지는 가능성을 감소
x_train = x_train/255.0
x_test = x_test/255.0

# 3) One-hot encoding (숫자를 나타내는 배열)
# The y_tain data is the associated labels for all the x_train examples. Rather
# than storing the label as an integer, it should be stored as a 1 * 10 binary
# array with the one representing the digit. This is known as _ one-hot encoding_.
y_train = to_categorical(y_train, num_classes=num_classes)
y_test = to_categorical(y_test, num_classes = num_classes)
# 벡터(정수)를 이진 클래스 행렬로 변환

# 하이퍼 파라미터 설정
# In computer programming it is generally best to use variables and constants
# rather than having to hard-code specific numbers every time that number is used.
epochs = 20
batch_size = 128
hidden_units = 32

# Step 2 : 모델 구성
model = Sequential([
                    Dense(hidden_units, activation = 'relu',input_dim= input_size),
                    Dense(10, activation = 'softmax')
])

# Step 3 : 모델 컴파일
model.compile(optimizer='rmsprop', 
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])

# Step 4 : 모델 학습
hist = model.fit(x_train, y_train, epochs=epochs, 
                 batch_size=batch_size,  validation_split=0.3, verbose = 1)
            # verbose = 0 (silent), 1 (progress bar), 2 (one line per epoch)
            # default = 1

# Loss와 Accuracy 시각화 
plot2(hist.history['loss'],  'train loss',hist.history['val_loss'],'validation loss', 'Loss','epochs','loss')
plot2(hist.history['accuracy'], 'train accuracy', hist.history['val_accuracy'],'validation accuracy', 'Accuracy','epochs','accuracy')

# Step 5: 모델 평가
model.evaluate(x_test, y_test, batch_size= batch_size) 

print('\nExecution time in seconds:', time.time()-start) #  실행시간 출력
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

plt.figure(figsize=(6, 6))

predicted_value = model.predict(x_test)

cm = confusion_matrix(np.argmax(y_test, axis = -1),
                      np.argmax(predicted_value, axis = -1))

sns.heatmap(cm, annot = True, fmt = 'd')

plt.show()

# Check accuracies among 10 groups (정확성 확인)
print(cm)
print('\n')

for i in range(10):
    print(('label = %d\t(%d/%d)\taccuracy = %.3f') %
          (i, np.max(cm[i]), np.sum(cm[i]),
           np.max(cm[i])/np.sum(cm[i])))
    
# Test Data에서의 숫자별 정확도 확인
label_distribution = np.zeros(10)
prediction_distribution = np.zeros(10)

for idx in range(len(y_test)):

    label = int(np.argmax(y_test[idx]))

    label_distribution[label] = label_distribution[label] + 1

    prediction = int(np.argmax(predicted_value[idx]))

    prediction_distribution[prediction] = prediction_distribution[prediction] + 1


print(label_distribution)
print(prediction_distribution)
```
