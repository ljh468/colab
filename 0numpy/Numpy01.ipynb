# **Iterating over array elements**

배열요소 반복

 python은 배열 반복이 느림<br>
 반복을 피해야하지만, 어쩔수없는경우
 for loop를 사용함

```python
v = array([1, 2, 3, 4])
for element in v :
    print(element)
```

```python
M = array([ [1, 2], [3, 4] ])
for row in M :
    print("row", row)
    for element in row:
        print(element)
```

```python
# 배열의 각 요소를 반복하고 수정해야할경우 for loop를 사용
for row_idx, row in enumerate(M):
    print('row_idx : ', row_idx, 'row : ', row)

    for col_idx, element in enumerate(row):
        print('col_idx : ', col_idx, 'element : ', element)
        # 행렬 매트릭스 : 각 자기자신 원소의 제곱
        M[row_idx, col_idx] = element **2
```

# **Vectorizing functions**

백터화 함수

좋은 성능을 얻기위해 벡터와 매트릭스를 반복하지않고<br>
벡터화 알고리즘을 사용해야한다<br>
스칼라 알고리즘을 벡터화 알고리즘으로 변환하는 첫 번째 단계는 작성하는 함수가 벡터 입력과 함께 작동하도록 하는 것임

```python
def Theta(x):
    # Scalar implemenation of the Heaviside step function.
    if x >= 0:
        return 1
    else:
        return 0
```

```python
%%script false
Theta(array([-3, -2, -1, 0, 1, 2, 3]))
```

```python
# Numpy 함수 벡터화를 사용
Theta_vec = vectorize(Theta)
type(Theta_vec)
```

```python
Theta_vec(array([-3, -2, -1, 0, 1, 2, 3]))
```

```python
# 스칼라에도 사용
Theta(-1.2), Theta(2.6)
```

# **Using arrays in conditions**

조건에서 배열사용

조건(예: 문 및 기타 부울식)에서 배열을 사용할 경우 일부 또는<br> 모두를 사용해야 합니다. 따라서 배열의 일부 또는 <br>모든 요소가 True로 평가되어야 합니다.

```python
M
```

```python
if (M > 5).any():
    print('at least one element in M is larger than 5')
    print('M에서 하나 이상의 원소가 5보다 큽니다.')
else:
    print('no element in M is larger than 5')
    print('M에 5보다 큰 원소가 없습니다.')
```

```python
if (M > 5).all():
    print('all elements in M are larger than 5')
    print('M의 모든 원소는 5보다 큽니다.')
else:
    print('all elements in M are not larger than 5')
    print('M의 모든 원소는 5보다 크지 않습니다.')
```
