# **Generative Adversarial Network (GAN)**

GAN은 두 개의 하위 네트워크(인코더와 디코더)가 서로 반대되는 손실 기능,<br> 즉 실제 데이터와 구별할 수 없는 데이터를 생성하도록 훈련된 인코더와<br> 데이터와 생성된 데이터를 구별하도록 훈련된 디코더에 대해 훈련된 일종의 신경 네트워크입니다.

생성적 적대 신경망 (Generative Adversarial Networks, GANs)은 요즘 컴퓨터 과학에서 가장 흥미로운 아이디어 중 하나입니다.<br> 두개의 모델이 적대적인 과정을 통해 동시에 훈련됩니다.<br> 생성자 (Generator, "모조품화가")는 진짜처럼 보이는 이미지를 생성하도록 배우는 와중에, 감별자 (Discriminator,"진품감정사")는 가짜의 이미지로부터 진짜를 구별하게 되는 것을 배우게 됩니다.

![생성자와 감별자를 그린 도표](https://tensorflow.org/tutorials/generative/images/gan1.png)


훈련과정 동안 *생성자*는 점차 실제같은 이미지를 더 잘 생성하게 되고, *Generateor*는 점차 진짜와 가짜를 더 잘 구별하게됩니다.<br> 이 과정은 DisCriminator가 가짜 이미지에서 진짜 이미지를 더이상 구별하지 못하게 될때, 평형상태에 도달하게 됩니다. 

![생성자와 감별자를 그린 두번째 도표](https://tensorflow.org/tutorials/generative/images/gan2.png)

__[GAN 쉽게 이해하기](https://doooob.tistory.com/172)__

> GAN(Generative Adversarial Networks)은 2개의 네트워크가 절차탁마하면서 
> 서로 성장해가는 비지도 학습 모델입니다. 이 두개의 네트워크는 각자의 역할이 있으며, 형 generator는 가짜 생성기이고, 동생 discriminator는 그것을 
> 간파하는 역할입니다. 이 관계에서 적대적(Adversarial)이라는 말이 붙었습니다만, 
> 실제로는 착한 동생이 형의 성장을 위해 협력하는 이미지입니다(GAN의 주역은 
> 형인 '생성기'입니다).   
>
> 형이 실물의 샘플을 모방한 가짜를 만들어서, 진짜와 가짜를 교대로 동생에게
> 보여줍니다. 동생은 가짜를 봤을 때, 어느 쪽의 확률이 높은지 판단하여,
> 시그모이드 함수(RNN때 나왔었죠)에 의해 사실(1)인지 가짜(0)인지 구별합니다.
> 그리고 동생이 정답 여부를 판정할 때, 오차 역전파(Backpropagation)에 의해 
> 형과 동생을 차례로 훈련(가중치 변경)시킵니다. 
>
> 일반적으로 훈련은 미니 배치 1세트씩 번갈아가며 수행하며, 형이 훈련할 때는 
> 동생을 고정하고 동생이 훈련할 때는 형을 고정함
>
> 이 훈련을 통해 형은 더욱 모방을 잘할 수 있게 되고, 동생도 뭐든 감정 가능한 
> 수준으로 성장합니다. 결국, 형이 "위작의 명인"이 되어 현실과 전혀 구분할 수
> 없는 가짜를 생성할 수 있게 되면, 동생의 판정 확률이 반반이 되면서 미션이 
> 종료됩니다.

# **GAN Model Architecture**

<img src="https://i.imgur.com/LweaD1s.png" width="600px">

- GAN 모델 아키텍처에는 두 개의 하위 모델이 포함됩니다.
    - Generator. 문제 도메인에서 타당성 있는 예를 새로 생성하는 데 사용되는 모델입니다.
    - Discriminator. 예를 실제(도메인에서) 또는 가짜(생성된)로 분류하는 데 사용되는 모델입니다.

- GAN은 발전기 네트워크가 적과 경쟁해야 하는 게임 이론 시나리오에 기반을 두고 있다.
    - 제너레이터 네트워크는 샘플을 직접 생성합니다.
    - 이것의 상대인 판별자 네트워크는 훈련 데이터에서 추출된 표본과 발전기에서 추출된 표본을 구별하려고 시도합니다.
    - 그들은 제로섬 게임을 하고 있다.

- GAN 아키텍처의 영리한 특성은 생성 모델의 교육이 지도 학습 문제로 프레임화된다는 것이지만, 생성 모델링은 비지도 학습 문제입니다.



GAN에는 크게 세 가지 범주가 있습니다.
- Unsupervised GANs : 생성기 네트워크는 랜덤 노이즈를 입력으로 사용하여 교육 데이터 세트에 나타나는 이미지와 매우 유사한 사실적인 이미지를 생성합니다.
    - ex: 원래의 GAN, DC-GAN(딥 컨볼루션 GAN), pg-GAN(GAN의 점진적 성장) 등

- Style-Transfer GANs - 이미지를 한 도메인에서 다른 도메인으로 변환합니다(예: 말에서 얼룩말, 스케치에서 컬러 이미지로 변환).
    - ex: CycleGAN 및 Pix2pix.
    - Conditional GANs - 피쳐에 대해 공동으로 학습하여 해당 피쳐에 조건부 이미지를 생성합니다(예: 특정 클래스의 인스턴스 생성).
    - ex: Conditional GAN, AC-GAN(Auxiliary classifier GAN), Stack-GAN, and BigGAN.

## How to train a GAN

 - Discriminator
<img src='https://developers.google.com/machine-learning/gan/images/gan_diagram_discriminator.svg'>

- discriminator는 생성기의 실제 데이터와 가짜 데이터를 모두 분류합니다.
- discriminator loss는 판별자가 실제 사례를 가짜 또는 가짜 인스턴스로 잘못 분류한 것에 대해 불이익을 줍니다.
- discriminator는 discriminator loss에서 판별자 네트워크를 통한 역전파를 통해 가중치를 갱신합니다.
- discriminator 학습 중에 generator는 훈련하지 않습니다.

- Discriminator를 사용하여 generator 학습
    - 랜덤 노이즈 표본 추출.
    - 표본 추출된 랜덤 노이즈에서 제너레이터 출력을 생성합니다.
    - 제너레이터 출력에 대한 판별기 "Real" 또는 "Fake" 분류를 가져옵니다.
    - 판별 분류에서 손실을 계산합니다.
    - 판별기와 생성기를 통해 역전파하여 Gradients를 얻습니다.
    - Gradients으로 제너레이터 가중치만 변경합니다.

## GAN Training

<img src='https://929687.smushcdn.com/2407837/wp-content/uploads/2020/11/keras_gans_steps.png?lossy=1&strip=1&webp=0'>

GAN에는 별도로 교육된 두 개의 네트워크가 포함되어 있으므로<br> GAN의 교육 알고리즘은 다음과 같은 두 가지 문제를 해결해야 합니다.

- GAN은 서로 다른 두 가지 교육(generator 및 discriminator)을 교묘하게 다뤄야 합니다.
- GAN 융합은 식별이 어렵습니다.





교차 학습

generator 와 discriminator는 서로 다른 교육 과정을 가지고 있습니다. 그럼 GAN 전체를 어떻게 훈련시키죠?

GAN 교육은 다음 기간 동안 진행됩니다.
- 판별기는 하나 이상의 시대에 대해 훈련합니다.
- 제너레이터는 하나 이상의 에포크로 작동합니다.
- 제너레이터 및 판별기 네트워크를 계속 교육하려면 1단계와 2단계를 반복합니다.




Note

1. 우리는 판별 훈련 단계 동안 발전기를 일정하게 유지합니다.
2. 마찬가지로, 우리는 발전기 교육 단계 동안 판별기를 일정하게 유지합니다.
3. GAN을 훈련할 때, 우리의 목표는 최소 손실액을 찾는 것이 아니라 두 가지 사이의 균형을 찾는 것입니다.
4. 평형을 찾는다는 개념은 서류상으로는 타당할 수 있지만, 일단 독자적인 GAN을 구현하고 교육하려고 하면 이것이 사소한 프로세스가 아니라는 것을 알게 될 것입니다.



Convergence

generator이 훈련과 함께 개선될수록, discriminator는 실제와 가짜를 쉽게 구별하지 못하기 때문에 discriminator 성능이 나빠집니다. <br>
generator가 완벽하게 성공하면 discriminator의 정확도는 50%입니다.<br>실제로, discriminator는 동전을 던져 예측합니다.

이러한 진행은 GAN 전체의 융합에 문제를 제기하며, 시간이 지남에 따라 차별적 피드백의 의미가 낮아집니다.<br>
GAN이 generator가 완전히 무작위 피드백을 제공하는 시점을 지나 훈련을 계속하면 generator가 정크 피드백에 대한 훈련을 시작하고 자체 품질이 저하될 수 있습니다.

## The loss function

<img src='https://editor.analyticsvidhya.com/uploads/68909architecture.PNG'>

- $G$: Generator 모델
- $D$: Discriminator 모델
- $z$: 랜덤 노이즈(고정 크기 입력 벡터)
- $x$: Real Image
- $G(z)$ Generator에서 생성된 이미지(가짜 이미지)
- $p_{data}(x)$ 실제 이미지의 확률 분포
- $p_z(z)$ 가짜 이미지의 확률 분포
- $D(G(z))$ 생성된 이미지가 입력일 때 식별자의 출력
- $D(x)$ 실제 이미지가 입력인 경우 식별자의 출력
<br><br>


제너레이터 모델과 판별기 모델 간의 싸움은 다음과 같이 수학적으로 표현될 수 있습니다.

<img src='https://editor.analyticsvidhya.com/uploads/26563gan-equation.png'>

방정식에서 알 수 있듯이, 생성기는 V(D, G)를 최소화하려고 하는 반면 판별기는 V(D, G)를 최대화하려고 합니다.<br> 다음 두 가지 용어를 모두 이해하겠습니다.

- $E_{x \sim p_{data}(x)} [\log D(x)]$: 실제 이미지가 입력되었을 때 D의 평균 로그 확률입니다.
- $E_{z \sim p_z(z)} [\log(1 – D(G(z)))]$: 생성된 이미지가 입력될 때 D의 평균 로그 확률입니다.

<hr>
<br>

**Discriminator 의 관점**

Discriminator는 실제 이미지와 가짜 이미지를 올바르게 분류하여 손실 함수 V(D,G)를 최대화하려고 합니다.

첫 번째 항은 Discriminator가 D(x)를 가능한 1에 가깝게 만들고자 한다는 것을 나타냅니다.<br> 즉, 실제 이미지를 실제 이미지로 올바르게 분류합니다.

두 번째 항은 Discriminator가 D(G(x))를 가능한 0에 가깝게 만들려고 한다는 것을 나타냅니다.<br> 즉, 가짜 이미지를 올바르게 분류하여 결국 항을 최대화하려고 합니다(1 – 숫자가 작을수록 숫자가 커짐). <br>참고: 확률은 0-1 범위에 있습니다.<br>

따라서 판별기는 두 항을 모두 최대화하려고 합니다.



<hr><br>

**Generator의 관점**

Generator는 실제 이미지처럼 보이는 이미지를 생성하여 손실 함수 V(D,G)를 최소화하려고 하며 Discriminator를 속이려고 합니다.

두 번째 항은 생성기가 가능한 한 D(G(z))를 0이 아닌 1에 가깝게 만들고,<br> 따라서 항을 최소화(1 – 숫자가 크면 숫자가 작아짐)하려고 한다는 것을 나타냅니다.<br> 식별자가 실패하여 이미지를 잘못 분류하도록 합니다.

따라서 Generator는 두 번째 항을 최소화하려고 합니다.

## Load and prepare the dataset

MNIST 데이터 집합을 사용하여 generator와 discriminator를 교육합니다. 생성기는 MNIST 데이터와 유사한 손으로 쓴 숫자를 생성합니다.

```python
(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()
print(train_images.shape)
print(train_labels.shape)
print(_.shape)
```

```python
train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')
print(train_images.shape)                      # 0 ~ 255의 정수 값일때
train_images = (train_images - 127.5) / 127.5  # 이미지를 [-1, 1]로 정규화합니다.
print(train_images[0][0])
```

```python
BUFFER_SIZE = 60000
BATCH_SIZE = 256
```

```python
# 데이터 배치를 만들고 섞습니다.
train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
```

# Create the models

generator와 discriminator는 모두 Keras [Keras Sequential API](https://www.tensorflow.org/guide/keras#sequential_model).를 사용하여 정의됩니다.


## The Generator

생성기는 `tf.keras.layers.Conv2DTranspose`(업샘플링) 레이어를 사용하여 시드(랜덤 노이즈)에서 이미지를 생성합니다.<br> 이 시드를 입력으로 사용하는  `Dense`레이어로 시작한 다음 원하는 이미지 크기인 28x28x1이 될 때까지 샘플을 여러 번 올립니다.<br>
tanh를 사용하는 출력 계층을 제외한 각 계층에 대한 `tf.keras.layers.LeakyReLU` 활성화가 나타납니다.

```python
def make_generator_model():
    model = tf.keras.Sequential()
    model.add(layers.Dense(7*7*256, use_bias=False, input_shape=(100,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((7, 7, 256)))
    assert model.output_shape == (None, 7, 7, 256)  # Note: None is the batch size

    model.add(layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False))
    assert model.output_shape == (None, 7, 7, 128)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert model.output_shape == (None, 14, 14, 64)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert model.output_shape == (None, 28, 28, 1)

    return model
```

```python
g = make_generator_model()
g.summary()
```

# The Disriminator

discriminator는 CNN 기반 이미지 분류자입니다.

```python
def make_discriminator_model():
    model = tf.keras.Sequential()
    model.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same',
                                     input_shape=[28, 28, 1]))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Flatten())
    model.add(layers.Dense(1))

    return model
```

```python
d = make_discriminator_model()
d.summary()
```

```python
plot_model(d)
```

# **Define the loss and optimizers**

두 모델에 대해 손실 함수 및 최적화자를 정의합니다.

```python
# This method returns a helper function to compute cross entropy loss
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True) # logit 변환(from_logits=True)은 [0, 1] 범위를 가지는 확률을 [-inf, inf] 범위를 가지는 값으로 변환해주는 변환
```

## Discriminator loss

이 방법은 판별기가 실제 이미지와 가짜 이미지를 얼마나 잘 구별할 수 있는지를 정량화합니다.<br> 실제 이미지에 대한 판별자의 예측을 1s 배열과 비교하고 가짜(생성된) 이미지에 대한 판별자의 예측을 0s 배열과 비교합니다

```python
def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss
```

## Generator loss

발전기의 손실은 그것이 얼마나 잘 판별자를 속일 수 있었는지 수치화한다.<br>직관적으로 제너레이터의 성능이 양호한 경우 판별기는 가짜 이미지를 실제(또는 1)로 분류합니다.<br> 여기서 생성된 이미지에 대한 판별기 결정을 1s 배열과 비교합니다.

```python
def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)
```

두 네트워크를 별도로 교육하기 때문에 판별기와 생성기 최적화기가 다릅니다.

```python
generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)
```

## Save checkpoints

이 노트북은 장시간 실행 중인 교육 작업이 중단되는 경우에도 도움이 될 수 있는 모델을 저장 및 복원하는 방법을 보여줍니다.

```python
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")
checkpoint = tf.train.Checkpoint(generator_optimizer=generator_optimizer,
                                 discriminator_optimizer=discriminator_optimizer,
                                 generator=g,
                                 discriminator=d)
```

# **Define the training loop**

```python
EPOCHS = 200
noise_dim = 100
num_examples_to_generate = 16

# You will reuse this seed overtime (so it's easier)
# to visualize progress in the animated GIF)
seed = tf.random.normal([num_examples_to_generate, noise_dim])
```

훈련 loop는 generator가 랜덤 시드를 입력으로 수신하는 것으로 시작됩니다. 그 시드는 이미지를 만드는 데 사용됩니다.<br> 그런 다음 discriminator를 사용하여 실제 이미지(교육 세트에서 그린 이미지)와 가짜 이미지(생성기에서 제작한 이미지)를 분류합니다.<br> 이러한 각 모델에 대해 손실이 계산되고 generator와 discriminator를 업데이트하는 데 사용됩니다.

```python
# Notice the use of `tf.function`
# This annotation causes the function to be "compiled".
@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      generated_images = g(noise, training=True)

      real_output = d(images, training=True)
      fake_output = d(generated_images, training=True)

      gen_loss = generator_loss(fake_output)
      disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, g.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, d.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, g.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, d.trainable_variables))
```

```python
def train(dataset, epochs):
  epoch=0
  for epoch in tqdm(range(EPOCHS), desc='Epoch 진행상황 => '):
    for image_batch in dataset:
      train_step(image_batch)

    # Produce images for the GIF as you go
    display.clear_output(wait=True)
    generate_and_save_images(g,
                             epoch + 1,
                             seed)

    # Save the model every 15 epochs
    if (epoch + 1) % 15 == 0:
      checkpoint.save(file_prefix = checkpoint_prefix)

    print ('Time for epoch {} is {} sec'.format(epoch + 1, time.time()-start))

  # Generate after the final epoch
  display.clear_output(wait=True)
  generate_and_save_images(g,
                           epochs,
                           seed)
```

Generate and save images

```python
def generate_and_save_images(model, epoch, test_input):
  # Notice `training` is set to False.
  # This is so all layers run in inference mode (batchnorm).
  predictions = model(test_input, training=False)

  fig = plt.figure(figsize=(4, 4))

  for i in range(predictions.shape[0]):
      plt.subplot(4, 4, i+1)
      plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
      plt.axis('off')

  plt.savefig('/content/images/image_at_epoch_{:04d}.png'.format(epoch))
  plt.show()
```

# **Train the model**

생성자와 구분자를 동시에 훈련시키려면 위에 정의된 '지우기' 방법을 호출하세요.<br> GAN을 훈련시키는 것은 까다로울 수 있습니다. 제너레이터와 판별기가 서로 위력을 발휘하지 않는 것이 중요합니다(예: 비슷한 속도로 훈련).

교육이 시작될 때 생성된 이미지는 랜덤 노이즈처럼 보입니다.<br> 교육이 진행됨에 따라 생성된 숫자가 점점 실제처럼 보일 것입니다.<br> 약 50개의 epoch 이후에는 MNIST 자리와 유사합니다.<br> 이 작업은 Collab의 기본 설정으로 약 1분/epoch가 소요될 수 있습니다.


```python
import time
start = time.time()
train(train_dataset, EPOCHS)
```

```python
print('Training Time in seconds: ', time.time()-start)
print('Training Time in mninutes: ', (time.time()-start)/60)
```
